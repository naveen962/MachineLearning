```python
#importing libraries
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd 
from sklearn import *
```

```python
#importing datasets  
data= pd.read_csv('bike_sharing.csv') 
data = data.drop(columns="dteday")
data
```

```python
X = np.array(data.drop(columns=['instant', 'season', 'yr', 'mnth', 'hr', 'holiday', 'weekday',
       'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed',
       'casual', 'registered'], axis=0))
Y = np.array(pd.DataFrame(data.cnt))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 90)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, Y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
# Visualizing the polynomial regression results
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X,Y, color ='red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color ='green')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
poly_pred=lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
print(poly_pred)
```
