```python
dataset.columns =  ['age', 'workarea', 'ID', 
           'education', 'education-year', 
           'marital-status', 'occupation', 
           'relationship', 'race', 'sex', 
           'capital-gain', 'capital-loss', 
           'hours-per-week', 'native-country', 
           'salary']
```

```python
dataset['salary'].replace([' <=50K',' >50K'],[0,1],inplace=True)
dataset
```

```python
X=dataset[['age','education-year','capital-gain','capital-loss','hours-per-week']]
Y=dataset['salary']
```

```python
def normalize(values):
    return (values - values.min())/(values.max() - values.min())
cols=['age','education-year','capital-gain','capital-loss','hours-per-week']
dataset[cols] = dataset[cols].apply(normalize)
X=dataset[cols]
```

```python
train_index = int(0.8 * len(X))
X_train, X_test = X[:train_index], X[train_index:]
Y_train, Y_test = Y[:train_index], Y[train_index:]
```

```python
a=0
b=0
c=0
d=0
e=0
f=0

#learning rate
L = 0.01 
#epoch for mininum iterations to find minimum rss/mse
epochs = 1000  

#number of observations
n = float(len(X_train)) 

for i in range(epochs): 
    z = a+b*X_train[:,0]+c*X_train[:,1]+d*X_train[:,2]+e*X_train[:,3]+f*X_train[:,4]
    h=1/(1+np.exp(-z))
    
    D_a = (1/n) * sum(h-Y_train)
    D_b = (1/n) * sum(X_train[:,0] * (h-Y_train))
    D_c = (1/n) * sum(X_train[:,1] * (h-Y_train)) 
    D_d = (1/n) * sum(X_train[:,2] * (h-Y_train))   
    D_e = (1/n) * sum(X_train[:,3] * (h-Y_train))
    D_f = (1/n) * sum(X_train[:,4] * (h-Y_train))
   
    a = a - L * D_a  
    b = b - L * D_b
    c = c - L * D_c  
    d = d - L * D_d
    e = e - L * D_e
    f = f - L * D_f
    
    
print (a,b,c,d,e,f)
```

```python
z1 = a+b*X_train[:,0]+c*X_train[:,1]+d*X_train[:,2]+e*X_train[:,3]+f*X_train[:,4]
h_result=1/(1+np.exp(-z1))
```

```python
y_pred=[]

for i in range(len(h_result)):
    if h_result[i]>0.5:
        y_pred.append(1)
    elif h_result[i]<0.5:
        y_pred.append(0)
```
